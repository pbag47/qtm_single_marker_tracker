```python
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from swarm_object_class import SwarmObject


SWARM_MANAGER = SwarmObject()

cf_colors = {'cf1': 'white',
             'cf2': 'cyan',
             'cf3': 'yellow',
             'cf4': 'magenta',
             'cf5': 'blue',
             'cf6': 'red',
             'cf7': 'green',
             'cf8': 'pink',
             'cf9': 'brown',
             'cf10': 'orange'}

figure = make_subplots(rows=1, cols=1,
                           specs=[[{'type': 'scatter'}]])

figure.update_layout(title_text='Crazyflie initial position', template='plotly_dark', legend_tracegroupgap=50)

for agent in agents:
    figure.add_trace(go.Scatter(x=[agent.initial_position[0]],
                                y=[agent.initial_position[1]],
                                line=dict(color=cf_colors[agent.name]),
                                name=agent.name + ' initial position',
                                legendgroup=agent.name,
                                mode='markers',
                                ),
                    )
    
    figure.add_trace(go.Scatter(x=[agent.initial_position[0] + SWARM_MANAGER.xy_auto_avoid_d0 * np.cos(np.pi/50 * t) for t in range(100)],
                                y=[agent.initial_position[1] + SWARM_MANAGER.xy_auto_avoid_d0 * np.sin(np.pi/50 * t) for t in range(100)],
                                line=dict(color=cf_colors[agent.name], dash='dash'),
                                name=agent.name + ' auto avoidance security area',
                                legendgroup=agent.name,
                                mode='lines',
                                ),
                    )

figure.show()
```
